### UNDERSAMPLING

```python
from imblearn.under_sampling import NearMiss
```

```python
from imblearn.under_sampling import NearMiss

nm = NearMiss(sampling_strategy="auto", version=1)  # version=1, 2, or 3
X_res, Y_res = nm.fit_resample(X, Y)
```

```python
X_res.shape,Y_res.shape
```

```python
from collections import Counter
print(Counter(Y))
print(Counter(Y_res))

```

here we observe the undersampling of the dataset where the sample with most 98 percent of the data was under sampled to match the undersampled data

1. Split into Train and Test Sets
Even though the data is balanced, we still split it into training and testing to evaluate how well the model performs on unseen data.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, Y_res, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred.mean()
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### OVERSAMPLING

Now we will perform oversamopling whihc is done by SMOTE 


```python
from imblearn.over_sampling import SMOTE
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y
)
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, Y_train)
```

```python
print("Before SMOTE:", Counter(Y_train))
print("After SMOTE:", Counter(y_resampled))
```

This clearly shows the sample have been scaled to the 227451 that was the 90 percent of imbalanced set

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

model = RandomForestClassifier(random_state=42)
model.fit(X_resampled, y_resampled)

# Evaluate on original test data (not resampled!)
```

```python
Y_pred = model.predict(X_test)
print(classification_report(Y_test, Y_pred))
```
